```python
import random
n = random.randint(10, 100)
count_five=''
count_three=''

if n % 5 == 0:
    count_five = n//5
elif n % 3 == 0:
    count_three = n//3
elif n % 5 < 3:
    count_five = n//5
    count_three = 1 
else:                  # 나머지가 3보다 클 때
    count_five = n//5
    count_three = (n%5)//3 +1
    
print(f'흙 {n}kg을 나르기 위해서는 3kg을 {count_three}번, 5kg을 {count_five}번 날라야 합니다. ')
```
